#### If we look at the 'genres' column it seems we need to clean this column. because it looks to have some good data contain inside.

#### Step 9 : Wrangling data in 'genres' column.

```python
genreId, name, url = [],[],[]

for genre in df0['genres']:
    genreId.append(genre[0]['genreId'])
    name.append(genre[0]['name'])
    url.append(genre[0]['url'])
df0_g = pd.DataFrame([genreId, name, url]).T
```

#### Result :

```python
df0_g
```

#### Rename columns and Reindex again !!

```python
df0_g.rename(columns={0:'genreId', 1:'type', 2:'url'}, inplace=True)
df0_g.index = df0_g.index + 1
```

#### Check 'df0_g' again

```python
df0_g
```

#### Combine 'type' column with DataFrame 'df1' as a 'df2'

### Remind : ****'df2' table we add new 'type' column.****  

```python
df2 = pd.concat([df1, df0_g['type']], axis=1, sort=False)
```

#### Result

```python
df2
```

#### You will see from the result that have 'type' column from table 'df0_g' joined to 'df2' table.
